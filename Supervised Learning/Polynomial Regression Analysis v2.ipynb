# Polynomial Regression Analysis
Polynomial regression is a form of regression analysis in which the relationship between the independent variable \(x\) and the dependent variable \(y\) is modeled as an nth degree polynomial.

The general form of the polynomial regression model is given by:
\[y = b_0 + b_1x + b_2x^2 + ... + b_nx^n + \epsilon\]
where:
- \(y\) is the dependent variable
- \(x\) is the independent variable
- \(b_0, b_1, ..., b_n\) are the coefficients
- \(\epsilon\) is the error term

Higher degrees of polynomial regression can capture more intricate patterns in the data but may also lead to overfitting, where the model fits the training data very well but performs poorly on unseen data. Therefore, it's essential to choose the degree of the polynomial carefully, often using methods like cross-validation to find the optimal balance between bias and variance.

Choosing the degree of polynomial regression when you have multiple features (also called multivariate polynomial regression) is not straightforward and depends on several factors, including the nature of your data and the complexity of the relationship you are trying to model.

Understand the Data: Plot your data (if possible) and look for patterns. If the relationships between the features and the target variable appear nonlinear, higher-degree polynomials might be appropriate.

Start Simple: Begin with a lower-degree polynomial (e.g., degree 2) and evaluate its performance. You can gradually increase the degree if the lower-degree model does not capture the complexity of the data well.

Cross-Validation: Use cross-validation to evaluate the performance of different polynomial degrees. This will help you determine which degree generalizes best to unseen data.

Watch Out for Overfitting: Higher-degree polynomials can fit the training data very well but may overfit, resulting in poor generalization to new data. Regularization techniques like Ridge or Lasso regression can help mitigate overfitting.

Cross-validation is a statistical method used to estimate the performance of a machine learning model on unseen data. It involves partitioning the data into multiple subsets, training the model on some of these subsets, and testing it on the remaining subsets. The goal is to ensure that the model generalizes well to new, unseen data, and to prevent overfitting.

The most commonly used form of cross-validation is k-fold cross-validation.

Data Partitioning: The original dataset is randomly partitioned into k equally sized subsets, or "folds".

Performance Aggregation: The performance metric (e.g., accuracy, mean squared error) is calculated for each of the k iterations.
The k performance metrics are then averaged to produce a single estimate of the model's performance.

When to use Polynomial Regression?

The choice between linear regression and polynomial regression depends on the nature of the relationship between the independent variable(s) and the dependent variable.

Linear Regression

Use linear regression when:

Linear Relationship:

The relationship between the independent variable(s) and the dependent variable is approximately linear. This can be checked visually using scatter plots or by examining the residuals.
Simplicity:

A simpler model is preferred for interpretation and communication. Linear models are easier to understand and explain compared to more complex models.
Good Performance:

Linear regression provides a good fit for your data and the model performs well on validation metrics. Sometimes, a simple linear model can be quite effective.
Example Use Cases:

Predicting housing prices based on square footage, number of bedrooms, etc., where the relationship is roughly linear.
Estimating the effect of advertising expenditure on sales revenue.

Polynomial Regression

Use polynomial regression when:

Non-Linear Relationship:

There is a clear, non-linear relationship between the independent variable(s) and the dependent variable. Polynomial regression can capture more complex patterns compared to linear regression.
Visual Inspection:

Visual inspection of the data (e.g., scatter plots) suggests a non-linear relationship that can be better captured with polynomial terms.
Improved Performance:

A polynomial model significantly improves performance metrics (e.g., lower mean squared error) compared to a linear model, indicating a better fit to the data.
Example Use Cases:

Modeling growth rates, where the relationship between time and growth is non-linear.
Predicting the effect of production costs on sales revenue, where the relationship is more complex than a straight line.

## Problem Statement
A technology company is planning to launch a new product. The company wants to understand the relationship between the revenue generated from sales and the production costs. The company's managers have noticed that the production costs and sales revenue change in a non-linear way. To understand this relationship and develop accurate pricing strategies, they want to use polynomial regression.


```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generating random data for production costs and sales revenue
np.random.seed(42)
production_costs = np.random.uniform(1000, 5000, 100)
sales_revenue = production_costs * 1.5 + np.random.normal(0, 500, 100) + production_costs**2 * 0.0005 + production_costs**3 * 0.0000002

# Creating a DataFrame for the data
data = pd.DataFrame({
    'Production Costs': production_costs,
    'Sales Revenue': sales_revenue
})
```

```python
# Splitting the data into features (X) and target (y)
X = data[['Production Costs']]
y = data['Sales Revenue']
```

```python
# Performing Polynomial Regression with degree 3 for more pronounced nonlinearity
# Polynomial regression can sometimes appear similar to linear regression if the polynomial degree is low or
# if the relationship between the variables isn't strongly nonlinear.

poly_features = PolynomialFeatures(degree=3)
X_poly = poly_features.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

# Making predictions
y_pred = model.predict(X_poly)

# Evaluating the model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

## Model Evaluation
- Mean Squared Error (MSE): `mse`
- R-squared (R2) Score: `r2`


```python
# Creating plots
plt.figure(figsize=(14, 6))

# Plotting the original data
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='Original Data')
plt.title('Production Costs vs Sales Revenue')
plt.xlabel('Production Costs')
plt.ylabel('Sales Revenue')
plt.legend()

# Plotting the polynomial regression curve
plt.subplot(1, 2, 2)
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label='Polynomial Regression Fit')
plt.title('Polynomial Regression Fit')
plt.xlabel('Production Costs')
plt.ylabel('Sales Revenue')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Creating an overview DataFrame to display the original and predicted values
results = pd.DataFrame({
    'Production Costs': production_costs,
    'Actual Sales Revenue': sales_revenue,
    'Predicted Sales Revenue': y_pred
})

# Display the DataFrame in Google Colab
results
```
