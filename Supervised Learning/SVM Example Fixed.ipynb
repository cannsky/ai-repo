Sonu√ß olarak, SVM algoritmasƒ± ile veri sƒ±nƒ±flandƒ±rma i≈ülemi ger√ßekle≈ütirilir. Bu y√∂ntem, √∂zellikle y√ºksek boyutlu ve karma≈üƒ±k veri k√ºmeleri i√ßin olduk√ßa etkilidir.

```python

# Gerekli k√ºt√ºphaneleri y√ºkleme
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Veri setini olu≈üturma
X, y = make_classification(n_samples=1000, n_features=20, n_informative=2, n_redundant=10, random_state=42)

# Veriyi eƒüitim ve test setlerine ayƒ±r
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Veriyi standardize et
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ƒ∞lk 5 √∂rneƒüe bakalƒ±m
pd.DataFrame(X_train).head()
```

```python

# SVM modelini eƒüitme
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svc = SVC(kernel='linear', random_state=42)
svc.fit(X_train, y_train)

# Test seti √ºzerinde tahmin yapma
y_pred = svc.predict(X_test)

# Doƒüruluk skorunu hesaplama
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Doƒüruluk Skoru: {accuracy:.2f}")
```

```python

# Modeli deƒüerlendirme
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred)
clr = classification_report(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(clr)
```

```python

# Sonu√ßlarƒ± g√∂rselle≈ütirme
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Tahmin Edilen Etiket")
plt.ylabel("Ger√ßek Etiket")
plt.show()
```

```python
# Gerekli k√ºt√ºphaneleri y√ºkleme
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns

# Veri setini olu≈üturma
X, y = make_classification(n_samples=1000, n_features=20, n_informative=2, n_redundant=10, random_state=42)

# Veriyi eƒüitim ve test setlerine ayƒ±r
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Veriyi standardize et
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ƒ∞lk 5 √∂rneƒüe bakalƒ±m
pd.DataFrame(X_train).head()

# SVM modelini eƒüitme
svc = SVC(kernel='linear', random_state=42)
svc.fit(X_train, y_train)

# Test seti √ºzerinde tahmin yapma
y_pred = svc.predict(X_test)

# Doƒüruluk skorunu hesaplama
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Doƒüruluk Skoru: {accuracy:.2f}")

# Modeli deƒüerlendirme
cm = confusion_matrix(y_test, y_pred)
clr = classification_report(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(clr)

# Sonu√ßlarƒ± g√∂rselle≈ütirme
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Tahmin Edilen Etiket")
plt.ylabel("Ger√ßek Etiket")
plt.show()

# 2 √∂zellik kullanarak SVM'i g√∂rselle≈ütirme
def plot_svc_decision_boundary(model, X, y, ax=None, plot_support=True):
    """Plot the decision boundary for a linear SVM"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # Create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = model.decision_function(xy).reshape(XX.shape)

    # Plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])

    # Plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=100,
                   linewidth=1, facecolors='none', edgecolors='k')
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

# ƒ∞ki √∂zellik se√ßelim (√∂rneƒüin, ilk iki √∂zellik)
X_vis = X[:, :2]

# ƒ∞ki √∂zellik ile model eƒüitimi
X_train_vis, X_test_vis, y_train_vis, y_test_vis = train_test_split(X_vis, y, test_size=0.3, random_state=42)
svc_vis = SVC(kernel='linear', random_state=42)
svc_vis.fit(X_train_vis, y_train_vis)

# Karar sƒ±nƒ±rlarƒ±nƒ± √ßizdirme
plt.figure(figsize=(10, 7))
plt.scatter(X_vis[:, 0], X_vis[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_boundary(svc_vis, X_vis, y)
plt.title("SVM Karar Sƒ±nƒ±rlarƒ± (2 √∂zellik ile)")
plt.xlabel("√ñzellik 1")
plt.ylabel("√ñzellik 2")
plt.show()
```
