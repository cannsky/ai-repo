### Hierarchical Clustering Example

This notebook demonstrates the Hierarchical Clustering algorithm using a synthetic dataset.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs

# Veri oluÅturma
X_hierarchical, y_hierarchical = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# HiyerarÅik KÃ¼meleme algoritmasÄ±nÄ± uygulama
linked = sch.linkage(X_hierarchical, method='ward')

# Dendrogram oluÅturma ve gÃ¶rselleÅtirme
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(linked)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()

# KÃ¼me sayÄ±sÄ±nÄ± belirleme ve kÃ¼meleme
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X_hierarchical)

# KÃ¼meleme sonuÃ§larÄ±nÄ± gÃ¶rselleÅtirme
plt.figure(figsize=(8, 6))
plt.scatter(X_hierarchical[:, 0], X_hierarchical[:, 1], c=y_hc, s=50, cmap='viridis')
plt.title("Hierarchical Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
